```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')

file_path_train = '/content/drive/MyDrive/MNIST/MNIST_train.csv'
file_path_test = '/content/drive/MyDrive/MNIST/MNIST_test.csv'

df = pd.read_csv(file_path_train)
df_test = pd.read_csv(file_path_test)
```

```python
# Features
X = df.iloc[:, 3:].to_numpy()
# Labels
y = df.iloc[:, 2].to_numpy()

# Features
X_test = df_test.iloc[:, 3:].to_numpy()
# Labels
y_test = df_test.iloc[:, 2].to_numpy()
```

```python
class KNN:
    def __init__(self, X_train, y_train):
        self.X = X_train
        self.y = y_train

    def predict(self, X, K, epsilon=1e-3):
        # X: shape (N_test, D)
        # self.X: shape (N_train, D)
        # Compute squared Euclidean distance efficiently using broadcasting:
        X_norm = (X ** 2).sum(axis=1).reshape(-1, 1)
        X_train_norm = (self.X ** 2).sum(axis=1).reshape(1, -1)
        dist2 = X_norm + X_train_norm - 2 * np.dot(X, self.X.T)

        # Get indices of K smallest distances
        idxt = np.argpartition(dist2, K, axis=1)[:, :K]

        y_hat = np.zeros(len(X))
        for i in range(len(X)):
            gamma_k = 1 / (np.sqrt(dist2[i, idxt[i]]) + epsilon)
            y_hat[i] = np.bincount(self.y[idxt[i]], weights=gamma_k).argmax()
        return y_hat
```

```python
# Initialize and test the KNN classifier
knn_classifier = KNN(X, y)
K = 5
y_pred = knn_classifier.predict(X_test, K)

# Evaluate the classifier
accuracy = np.mean(y_pred == y_test)
print(f"Accuracy of KNN classifier with K={K}: {accuracy}")
```
